```python
import numpy as np
import random
```

```python
def generate_circle(rad, points, inside=True):
    X = np.arange(points,dtype=float)
    Y = np.arange(points,dtype=float)
    c = np.zeros(points)
    for point in range(points):
        if inside:
            X[point] = random.uniform(-1,1)*rad
            Y[point] = random.uniform(-1,1)*rad
            c[point] = 1
        else:
#             X[point] = random.randint(-3*rad,3*rad) + 
#             Y[point] = random.randint(-3*rad,3*rad)
            c[point] = 0
            temp = random.uniform(-1,1)*rad* 3
            if temp > 0:
                X[point] = temp + rad
            else:
                X[point] = temp - rad
                
            temp = random.uniform(-1,1)*rad * 3
            if temp > 0:
                Y[point] = temp + rad
            else:
                Y[point] = temp - rad           
    
    return X,Y,c
```

```python
X1,Y1,c1 = generate_circle(6,100,False)
```

```python
X2,Y2,c2 = generate_circle(5,50, True)
```

```python
plt.scatter(X1,Y1,color='b')
plt.scatter(X2,Y2,color='g')
```

```python
x=np.arange(-30,30)
y=[]
for i in range(60):
    if i<15:
        y.append(1)
    elif i>=15 and i<=45:
        y.append(0)
    else:
        y.append(1)
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(np.array(x).reshape(-1,1),np.array(y).reshape(-1,1),test_size=0.2)
```

```python
from sklearn.svm import SVC 
clf = SVC(kernel='poly')
```

```python
clf.fit(x_train,y_train) 
```

```python
y_pred=clf.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
